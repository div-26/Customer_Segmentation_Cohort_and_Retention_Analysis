```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("2_cohort_rev_per_customer.csv")
df
```

```python
sns.set_theme(style= 'ticks')
sns.barplot(data= df, x= 'cohort_year', y = 'avg_revenue_per_customer', hue = 'total_customers')
sns.despine()
plt.plot(df.index, df['avg_revenue_per_customer'], color="#4A379F", marker='o', markersize = 4, linewidth=2)

plt.title("Average Revenue Per Customer By Cohort Year")
plt.legend().remove()
plt.ylabel('Average Revenue Per Customer', fontstyle = 'italic')
plt.xlabel('')

plt.figtext(0.5, -0.05,
    "*Darker bars indicate higher customer volume in that cohort year",
    ha='center',
    fontsize=9,
    fontstyle = 'italic'
)

ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, fig : 0 if y == 0 else f'{y/1000}K'))
```
