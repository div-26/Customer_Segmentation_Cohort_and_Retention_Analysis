```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("1_customer_segmentation.csv")
df
```

```python
df = df.set_index('customer_segment')

# Clean index labels by removing leading numbers and dashes (e.g., "1 - Low Value" → "Low Value")
df.index = df.index.str.replace(r'^\d+\s*-\s*', '', regex=True)

df
```

```python
df['segment_total_ltv'].plot(kind= 'pie', startangle = 90, autopct = '%1.1f%%')
plt.title('Total Lifetime Revenue per Segment')
```
