```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("3_retention_vs_churn.csv")
df
```

```python
sns.barplot(data= df, x= 'cohort_year', y = 'status_percentage', hue= 'customer_status', palette={'Active Customer': "#038ac8" , 'Churned Customer': "#ff9063"})
sns.despine()

ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, fig : f'{round(y)}%'))

plt.title('Customer Status Percentage By Cohort Year')
plt.ylabel('Percentage of customers')
plt.yticks(np.arange(0, 101, 10))
plt.xlabel('')
plt.grid(axis='y', linestyle='--', alpha = 0.5)
plt.legend(title = 'Customer Status', bbox_to_anchor = (1,0.2))
```
